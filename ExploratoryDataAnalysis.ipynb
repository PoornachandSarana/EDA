# Converting Categorical features into numerical values

```python
car1['EngineCC'].head()
```

```python
car1['MaxPower']= car1['max_power'].apply(lambda x : x.split(' '))
car1['MaxPower']=car1['MaxPower'].apply(lambda x: x[0])
car1['MaxPower']=car1['MaxPower'].astype('float64')
```

```python
car1['fuel'].unique()
```

```python
def fuel_val(x):
    if x=='Diesel':
        return 1
    elif x=='Petrol':
        return 2
    elif x=='LPG':
        return 3
    elif x=='CNG':
        return 4
```

```python
car1['fueltype']= car1['fuel'].map(fuel_val)
```

```python
car1['seller_type'].unique()
```

```python
def sellertype(x):
    if x=='Individual':
        return 1
    elif x=='Dealer':
        return 2
    elif x=='Trustmark Dealer':
        return 3
```

```python
car1['sellertype']=car1['seller_type'].map(sellertype)
```

```python
car1['transmission'].unique()
```

```python
def Transmission(x):
    if x=='Manual':
        return 1
    elif x=='Automatic':
        return 2
```

```python
car1['transtype']=car1['transmission'].map(Transmission)
```

```python
car1['owner'].unique()
```

```python
def own(x):
    if x=='First Owner':
        return 1
    elif x=='Second Owner':
        return 2
    elif x=='Third Owner':
        return 3
    elif x=='Fourth & Above Owner':
        return 4
    else:
        return 5
```

```python
car1['ownertype']=car1['owner'].map(own)
```
